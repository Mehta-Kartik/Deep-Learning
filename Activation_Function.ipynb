```python
import math
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(-5)
```

for x=-5 our sigmoid gives values close to zero

```python
sigmoid(100)
```

```python
sigmoid(10000000)
```

```python
sigmoid(4)
```

Now Let's try tanh function

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-1)
```

```python
tanh(0)
```

```python
tanh(1)
```

Now Let's try Relu

```python
def relu(x):
    return max(0,x)
```

```python
relu(-1)
```

```python
relu(1)
```

Now let's try leaky Relu

```python
def lrelu(x):
    return max(0.1*x,x)
```

```python
lrelu(-1)
```

```python
lrelu(5)
```

For -ve answer would be 0.1*x and for positive it would be x
