```python
import numpy as np
```

```python
ypred=np.array([1,1,0,0,1])
ytrue=np.array([0.3,0.7,1,0,0.5])
```

```python
def mae(ypred,ytrue):
    # return ((1/len(ypred))*np.sum(abs(ypred-ytrue))) This is shorter way we can also try another way
    totalloss=0
    for x,y in zip(ypred,ytrue):
        totalloss+=abs(x-y)
    mae=totalloss/len(ypred)
    return mae
```

```python
mae(ypred,ytrue)
```

```python
def mse(ypred,ytrue):
    # Shorter way
    return np.sum((ypred-ytrue)**2)/len(ypred) 

    #Longer Way
    # totalloss=0
    # for x,y in zip(ypred,ytrue):
    #     totalloss+=(x-y)**2
    # return totalloss/len(ypred)
mse(ypred,ytrue)
```

```python
def logloss(ypred,ytrue):
    # return (-np.sum((ytrue*np.log(ypred))+((1-ytrue)*np.log(1-ypred)))/len(ypred))
    return -np.mean(ytrue * np.log(ypred) + (1 - ytrue) * np.log(1 - ypred))
logloss(ypred,ytrue)
```

The above error occurred because log(0) is -inf which is not valid for any kind of operation
Here We would be trying to replace 1 with 0.99999999 and 0 with 0.000000000001 so that actual meaning of the value does not change nor we get error on log(0) so yes let's begin

```python
yprednew=[max(i,1e-15) for i in ypred]# We worked with replacing 0 here not its turn for 1
yprednew=[min(i,1-1e-15) for i in yprednew]
```

```python
yprednew=np.asarray(yprednew)# We also need to convert the list into np array because that property would have been lost in above line
```

```python
logloss(yprednew,ytrue)
```
